### 3. Filtering by using loc and iloc 
- We can filter the dataframe by using loc and iloc indexers as well

```python
# We can filter the DataFrame by using loc and iloc indexers as well
import pandas as pd
df=pd.read_csv('/Users/apple/Downloads/bank_transactions.csv')
#print(df)
con1=df["CustGender"]=="F"
con2=df["TransactionDate"]=="18/9/16"
df2=df.loc[con1 & con2]
print(df2)
```

```python
# Filtering DataFrame by using iloc indexer
import pandas as pd
df=pd.read_csv('/Users/apple/Downloads/bank_transactions.csv')
df1=df.iloc[:5,]
print(df1)
```

# and operator ==> &
# or  operator ==> |

```python
# Filtering DataFrame by using loc indexer
import pandas as pd
df=pd.read_csv('/Users/apple/Downloads/bank_transactions.csv')
con1=df["TransactionDate"]=="18/9/16"
con2=df["CustLocation"]=="NEW DELHI"
df1=df[con1 & con2]     # with And operater 
print(df1)
```

```python
# Filtering DataFrame by using loc indexer
import pandas as pd
df=pd.read_csv('/Users/apple/Downloads/bank_transactions.csv')
con1=df["TransactionDate"]=="18/9/16"
con2=df["CustLocation"]=="NEW DELHI"
df1=df[con1 | con2]     # with or operater 
print(df1)
```

### 3.3 isin() method
- isin() is predefines method in Series class.
