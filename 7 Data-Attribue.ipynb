### 1.11 T
- T is predefined attribute in DataFrame class
- We can access T attribute by using DataFrame object
- T means its Transpose, it returns rows as coloumn as row 

```python
import pandas as pd
details=[
    ["kishan",10000],
    ["firoj",20000],
    ["raja",40000],
    ["pika",90000]
]
c=["Name","salary"]
df=pd.DataFrame(details,columns=c)
df2=pd.read_csv("/Users/apple/Downloads/Walmart_Sales.csv")
df3=pd.read_csv('/Users/apple/Downloads/amazon.csv')
print(df)
print()
print("=> T attribute")
print(df.T)
print("*"*100)
print()
print(df2)
print()
print("=> T attributes ")
print(df2.T)
print()
print("*"*100)
print(df3)
print("=>  T attributes ")
print(df3.T)
```
