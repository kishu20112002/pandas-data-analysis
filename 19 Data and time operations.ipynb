| code         | description          |       Example                    |
|--------------|----------------------|----------------------------|
| &Y           | full year            |   2001                 |
| %m           | month w / Zero paddi  |  04                   |          | |%d            | Day of the month w/ zero padding  |  09       |
| %I           | Hour (12 hr clock) w/ Zero padding |02        |
| %p           | AM or PM                       | Am           |
| %M           | Minute w/ Zero padding    | 05 v              |
| %S           | Secounf W/ Zero padding    | 05  |            |           |                                                       |
|              |                      |                          |

### 8 Encoding Days of this week
- we can get the day of the week for each date by using pandas
  1. knowing the days names will helpful to understand the business flow. like we can compare total sales on specific day for the past three years

```python
# weeks name 
import pandas as pd
data={
    "Product":["Samsung","iphone","Motorola"],
    "Status":["Success","Success ","Failed"],
    "Cost":[10000,50000,15000],
    "Purdate":["10-oct-2025","11-oct-2025","12-oct-2025"]
}
df=pd.DataFrame(data)
df["Purdate"]=pd.to_datetime(df["Purdate"])
print(df)
print()
print(df["Purdate"].dt.day_name())
```
